Rock paper scissors.
encrypted strategy 
first column is what opponent plays
A for rock
B for paper
C for scissors
Second column is what you should play in response
X for Rock
Y for Paper
Z for scissors
Your score is the sum of your scores fr each round. 

```python
import numpy as np
import pandas as pd

# Define rock paper scissor scores
r1 = 0
p1 = 6
s1 = 3

# create a 2d mapping scores to hands
play = np.array([r1, p1 ,s1], dtype=str)
pset1 = np.roll(play, shift=-1)
pset2 = np.roll(pset1, shift=-1)
pset3 = np.roll(pset2, shift=-1)
table = pd.DataFrame([pset1, pset2, pset3], columns=['A', 'B', 'C'])
table.index = ['Y', 'X', 'Z']

# create a kv mapping for the actual hand values
hand_values = {"Y" : 2, "X":1, "Z":3}


def calulate_play(p1, p2):
  # given two hands calulate the score of the outcome (w, d, l) and the value of the hand
  outcome_value = table[p1][p2]
  hand_value = hand_values[p2]
  return int(outcome_value) + int(hand_value)

# sum the input
count = 0
for v in input:
  count += calulate_play(v[0], v[1])
print(count)
```
